# Correlación

```r
library(readxl)
library(ggplot2)
library(GGally)
library(corrplot)
library(PerformanceAnalytics)

# 2000
lyears = 2000:2020
ltmp = c()
lco = c()
lnox = c()
lpm = c()
lso2 = c()

for(year in lyears)
{
    CO <- read.csv(paste0("data/transfData/",year,"_CO.csv"))
    NOX <- read.csv(paste0("data/transfData/",year,"_NOX.csv"))
    SO2 <- read.csv(paste0("data/transfData/",year,"_SO2.csv"))
    TMP <- read.csv(paste0("data/transfData/",year,"_TMP.csv"))
    #PM2 <- read.csv(paste0("data/transfData/",year,"_PM2.5.csv"))
    
    lco = c(lco,summarise(CO,avg=mean(mean))[[1]])
    lnox = c(lnox,summarise(NOX,avg=mean(mean))[[1]])
    lso2 = c(lso2,summarise(SO2,avg=mean(mean))[[1]])
    ltmp <- c(ltmp,summarise(TMP,avg=mean(mean))[[1]])
    #lpm <- c(ltmp,summarise(PM2,avg=mean(mean))[[1]])
}

png("./figs/CORR.png",
    width=2800,height=1800,units="px",
    pointsize=12,bg="white",res=300)

base <- data.frame(lco,lnox,lso2,ltmp)
names(base) <- c("CO","NOX","SO2","TMP")
chart.Correlation(base, histogram = F, pch = 19)
```
